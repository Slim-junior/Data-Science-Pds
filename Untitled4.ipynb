```python
sets =pd.read_csv('sets.csv')
```

```python
sets.head()
```

```python
earliest_year = sets['year'].min()
first_year = sets[sets['year'] == earliest_year]
print(f"The first lego sets where released in:{earliest_year}.")
print(f"These sets where called")
print(first_year[['set_num', 'name']])
```

```python
sets.sort_values('num_parts', ascending=False).head()
```

```python
import matplotlib.pyplot as plt
```

```python
sets_by_year = sets.groupby('year').count()
sets_by_year['set_num'].head()
```

```python
plt.plot(sets_by_year.index, sets_by_year.set_num)
```

```python
plt.plot(sets_by_year.iloc[:-2].index, sets_by_year.iloc[:-2].set_num)
# Explanation: The 'index' here refers to the 'year' column of the sets_by_year DataFrame.
# When sets are grouped by year, the year becomes the index of the resulting DataFrame.
# So, sets_by_year.index represents the years (e.g., 1949, 1950, 1953, etc.) on the x-axis of the plot.
```
