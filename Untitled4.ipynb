```python
sets_by_year = sets.groupby('year').count()
sets_by_year['set_num'].head()
```

```python
plt.plot(sets_by_year.index, sets_by_year.set_num)
```

```python
plt.plot(sets_by_year.iloc[:-2].index, sets_by_year.iloc[:-2].set_num)
# Explanation: The 'index' here refers to the 'year' column of the sets_by_year DataFrame.
# When sets are grouped by year, the year becomes the index of the resulting DataFrame.
# So, sets_by_year.index represents the years (e.g., 1949, 1950, 1953, etc.) on the x-axis of the plot.
```

```python
themes_as_year = sets.groupby('year').agg({'theme_id':pd.Series.nunique})
themes_as_year.rename(columns={'theme_id':'nr_themes'},inplace=True)
themes_as_year.head()
```

```python
plt.plot(themes_as_year.iloc[:-3].index, themes_as_year.iloc[:-3].nr_themes)
```

```python
plt.plot(themes_as_year.iloc[:-3].index, themes_as_year.iloc[:-3].nr_themes)
plt.plot(themes_as_year.iloc[:-3].index, themes_as_year.iloc[:-3].nr_themes)
```

```python
ax1 = plt.gca() #get axes
ax2 = ax1.twinx()

ax1.plot(sets_by_year.iloc[:-2].index, sets_by_year.iloc[:-2].set_num,'pink')
ax2.plot(themes_as_year.iloc[:-3].index, themes_as_year.iloc[:-3].nr_themes, color='orange')
```

# LEGO Data Analysis

## Project Overview
This project involves an exploratory data analysis (EDA) of LEGO datasets, including information about LEGO colors and sets. The analysis focuses on understanding the distribution of colors, the history of LEGO set releases, and trends in the number of sets and themes over time.

## Data Sources
- `colors.csv`: Contains information about LEGO colors, including their names, RGB values, and transparency status.
- `sets.csv`: Contains details about individual LEGO sets, such as set number, name, year of release, theme ID, and number of parts.
- `themes.csv`: Provides details about LEGO themes, linking theme IDs to their names and parent themes.

## Key Findings
- **Color Analysis**: Explored the variety and distribution of LEGO colors, including the count of unique colors and the proportion of transparent colors.
- **Set Release Trends**: Visualized the number of LEGO sets released per year, identifying periods of significant growth and change in production.
- **Theme Evolution**: Analyzed the evolution of LEGO themes over time, observing how the diversity of themes has changed annually.
- **Largest Sets**: Identified the largest LEGO sets based on the number of parts.

## Tools and Libraries
- Python 3
- Pandas for data manipulation and analysis.
- Matplotlib for data visualization.
- Scikit-learn (used in a separate demonstration for a simple ML model, not directly part of the LEGO EDA).
