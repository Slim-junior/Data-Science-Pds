# Google Play Store App Analytics

This notebook performs an initial exploration and cleaning of a dataset containing Google Play Store app information. The goal is to understand the data, identify potential issues, and prepare it for further analysis.

## 1. Data Loading and Initial Inspection

*   **Libraries Imported**: `pandas`, `numpy`, `matplotlib.pyplot`, `plotly.express`, `seaborn`.
*   **Dataset**: The `apps.csv` file was loaded into a pandas DataFrame named `df_apps`.
*   **Initial Checks**:
    *   `df_apps.shape`: Confirmed the dataset has 10841 rows and 12 columns.
    *   `df_apps.sample()`, `df_apps.head()`, `df_apps.sample(5)`: Reviewed the first few rows and random samples to get a feel for the data structure and content.

## 2. Data Cleaning and Preprocessing

### Handling Missing Values

*   **Identifying Missing Ratings**: Rows with `NaN` values in the 'Rating' column were identified (`man_rows`). There were 1474 such rows.
*   **Dropping Missing Values**: All rows containing any `NaN` values were dropped, creating a new DataFrame `df_apps_clean`. The shape was reduced to 9365 rows and 12 columns.

### Handling Duplicates

*   **Identifying Duplicates**: Duplicated rows in `df_apps_clean` were found, totaling 474 rows.
*   **Investigating Specific Duplicates**: Checked for duplicates of popular apps like 'Facebook' and 'Instagram' to understand the nature of the duplications.
*   **Removing Duplicates**: Duplicates were removed based on the combination of 'App', 'Type', and 'Price' columns to ensure each unique app offering (by type and price) is represented only once. This further refined `df_apps_clean`.

### Data Type Conversion

*   **Installs Column**:
    *   Initially, the 'Installs' column was an object type and contained commas (e.g., '1,000,000').
    *   Commas were removed, and the column was converted to a numeric type.
    *   Grouped by 'Installs' to show the distribution of install counts.
*   **Price Column**:
    *   Initially, the 'Price' column was an object type and contained '$' (e.g., '$13.99').
    *   The '$' symbol was removed, and the column was converted to a numeric type.

### Filtering Data

*   **Filtering by Price**: Apps with a 'Price' greater than or equal to $200 were removed from `df_apps_clean` to focus on more reasonably priced applications.

## 3. Initial Data Exploration

*   **Top Rated Apps**: Sorted `df_apps_clean` by 'Rating' in descending order to identify the highest-rated apps.
*   **Largest App Sizes**: Sorted `df_apps_clean` by 'Size_MBs' in descending order to find the apps with the largest file sizes.
*   **Most Reviewed Apps**: Sorted `df_apps_clean` by 'Reviews' in descending order to see which apps have the most user reviews.
*   **Content Rating Distribution**:
    *   Calculated the value counts for the 'Content_Rating' column.
    *   Visualized the distribution of content ratings using two `plotly.express.pie` charts: one standard and one with a `hole` (donut chart style).
*   **Installs Distribution**: Examined the unique values and their counts in the 'Installs' column after cleaning and conversion.
*   **Price Distribution**: Investigated the 'Price' column after cleaning and conversion, specifically looking at the most expensive apps remaining after the filtering step.

This cleaned and preprocessed `df_apps_clean` DataFrame is now ready for more in-depth analysis and visualization of trends within the Google Play Store data.
